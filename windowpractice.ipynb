```python
#software making program:

def adhar_updation():
    #Adhar card updation center:
    #welcome student
    print("welcome adhar card center".title())
    #ask name
    name=input("please enter your name here:")
    print("your name saved successfully",name)
    #ask its phone no.
    ph_no=input("please enter your phone number here:")
    print("phone number saved successfully",ph_no)
    #ask its email id
    email=input("please enter your Email ID here:")
    print("Email ID saved successfully",email)
    #ask its course
    add=input("please enter your address here:")
    print("your address saved successfully",add)
    #ask its location
    location=input("enter your location:")
    print("your location will be",location)





def Travel_diary():
    #mumbai travel diary:
    #welcome user
    print("'welcome mumbai travel diary'".title())
    #ask user name
    user_name=input("please enter your name here:")
    #greet user
    print("hello!",user_name.title())
    #ask user to its budget
    user_budget=int(input("please enter your budget:"))
    #5000>visit taj hotel
    if(user_budget>5000):
        Print("please visit 'Taj hotel'")
    #3000-5000 visit get way of india
    elif(user_budget>3000 and user_budget<=5000):
        print("please visit 'Get way of india'")
    #2000-3000 visit neheru centre
    elif(user_budget>2000 and user_budget<=3000):
        print("please visit 'Neheru centre'")
    #1000-2000 visit jijabai bhosale udyan
    elif(user_budget>1000 and user_budget<=1000):
        print("please visit 'Jijabai bhosale udyan'")
    #less than 500 no option
    else:
        print("we don't have option for your budget")
        print("Thank you for visiting here!")
        print("have a nice day".title())




def multi_generator():    
    #welcome to multiplication game:
    print("welcome to multiplication game".upper())
    #user input:
    x=int(input("please enter a number here="))
    for i in range(1,11):
        print(x,"x",i,"=",x*i)



        

def cubic_game():
    #welcome to cubic game:
    print("welcome to cubic game".upper())
    #ask the input:
    x=int(input("please enter a number here="))
    for x in range(1,x+1):
        print("cube of",x,"=",x**3)




def factorial():    
    #factorial:
    #welcome to factorial game:
    print("welcome to factorial game".upper())
    x=1
    #ask the input:
    i=int(input("please enter a number here="))
    for i in range(1,i+1):
        x=x*i
    print("factors of",i,"=",x)




def fibbo_game():    
    #fibo number generrator:
    #welcome user:
    print("welcome to fibo generator".upper())
    n0=0
    n1=1
    #ask the input:
    x=int(input("please enter a number="))
    for x in range(1,x+1):
        print(n0)
        n2=n0+n1
        n0=n1
        n1=n2




def heads_tails():
    #heads and tails game:
    #welcome user:
    print("welcome heads and tails game".title())
    #ask for heads and tails:
    user_choice=input("please chose Heads and Tails:").title()
    #display user choose:
    print("You chose:",user_choice)
    #coin toss:
    import random
    if random.choice("ht")=="h":
        coin_toss="Heads"
    else:
        coin_toss="Tails"
    #display coin result:
    print("Coin Result:",coin_toss)
    #if user choice=coin result-->you win
    if user_choice==coin_toss:
        print("You win")
    #else you lose!
    else:
        print("You lose")




def dice_game():   
    #dice game:
    #welcome user:
    print("welcome to dice game".title())
    #ask for dice roll:
    user_choice=input("please enter a number between 1 to 6 here:")
    #display user choose:
    print("You chose:",user_choice)
    #roll dice:
    import random
    dice_result=random.choice("123456")
    #display dice result:
    print("Dice Result",dice_result)
    #if user choice=coin result-->you win
    if user_choice==dice_result:
        print("You win")
    #else you lose!
    else:
        print("You lose")




def double_dice():
    #double dice game:
    #welcome user:
    print("welcome to dice game".title())
    #ask for dice roll:
    user_choice=input("please enter a number between 2 to 12 here:")
    #display user choose:
    print("You chose:",user_choice)
    #roll dice:
    import random
    dice_result=random.randint(2,12)
    #display dice result:
    print("Dice Result:",dice_result)
    #if user choice=coin result-->you win
    if user_choice==dice_result:
        print("You win")
    #else you lose!
    else:
        print("You lose")





def password():    
    #Password Generator:
    import random
    import string
    user_input=int(input("enter number of character:"))
    if user_input>8:
        lower=string.ascii_lowercase
        upper=string.ascii_uppercase
        number=string.digits
        punc=string.punctuation
        #2 sample each
        lower_2_char=random.sample(lower,2)
        upper_2_char=random.sample(upper,2)
        number_2_char=random.sample(number,2)
        punc_2_char=random.sample(punc,2)
        basic_pass_list=lower_2_char+upper_2_char+number_2_char+punc_2_char
        remaining_len=user_input-8
        for i in range(remaining_len):
            basic_pass_list.append(random.choice("abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*()"))
        random.shuffle(basic_pass_list)
        password=" ".join(basic_pass_list)
        print("here is your password:",password)
    else:
        print("please use more number of characters for a strong password!")
       




def countdown():    
    from countdown import countdown
    user_min=int(input("enter minutes here:" ))
    user_sec=int(input("enter seconds here: "))
    countdown(mins=user_min,secs=user_sec)





def quiz_game():    
    #QUIZ GAME:
    #WELCOME USER
    print("welcome to QUIZ GAME!")
    #ASK USER NAME
    x=input("please enter your name:")
    #GREET USER
    print("Hello",x.title())
    #INITIALIZE SCORE BOARD
    score=0
    right_Q=0
    wrong_Q=0
    attempt_Q=0
    #ASK THE QUESTION
    print("Q.who is the capital of india?")
    print("And here is your option-A:delhi,B:mumbai,C:kolkata")
    ans=input("please enter your right option here:").upper()
    
    #GIVE THE SCORE ACCORDING TO ANSWER
    if ans=="A":
        print("correct answer")
        score+=10
        right_Q+=1
        attempt_Q+=1
    else:
        print("Oops!")
        score=0
        wrong_Q+=1
        attempt_Q+=1
    #ASK THE QUESTION
    print("Q.In which city jijabai bhosale udyan loacted ?")
    print("And here is your option-A:delhi,B:mumbai,C:kolkata")
    ans=input("please enter your right option here:").upper()
    #GIVE THE SCORE ACCORDING TO ANSWER
    if ans=="B":
        
        print("correct answer")
        score+=10
        right_Q+=1
        attempt_Q+=1
    else:
        print("Oops!")
        score=0
        wrong_Q+=1
        attempt_Q+=1
        
    #FINAL SCORE BOARD
    print("Attempt question=",attempt_Q)
    print("Right answer=",right_Q)
    print("Wrong answer=",wrong_Q)
    print("Final score=",score) 




#creating window:

import tkinter as tk
window=tk.Tk()
#window title:
window.title("MY Project")
#window size:
window.geometry("500x500")
#window label:
tk.Label(window,text="PYTHON PROJECT",font=("Helvetica",25)).place(x=110,y=30)
tk.Label(window,text="Made by : Yogita Aghav",font=("Helvetica",14)).place(x=170,y=80)
#window button:
tk.Button(window,text="   Adhar Updation    ",command=adhar_updation,font=("Helvetica",15)).place(x=40,y=150)
tk.Button(window,text="   Multi Generator    ",command=multi_generator,font=("Helvetica",15)).place(x=40,y=203)
tk.Button(window,text="     Travel Diary       ",command=Travel_diary,font=("Helvetica",15)).place(x=40,y=253)
tk.Button(window,text="    Heads & Tails     ",command=heads_tails,font=("Helvetica",15)).place(x=40,y=303)
tk.Button(window,text="     Double Dice       ",command=double_dice,font=("Helvetica",15)).place(x=40,y=353)
tk.Button(window,text="     Quiz Game        ",command=quiz_game,font=("Helvetica",15)).place(x=40,y=403)

tk.Button(window,text="     Fibbo Game     ",command=fibbo_game,font=("Helvetica",15)).place(x=270,y=150)
tk.Button(window,text="     Cubic Game     ",command=cubic_game,font=("Helvetica",15)).place(x=270,y=203)
tk.Button(window,text="     Dice Game       ",command=dice_game,font=("Helvetica",15)).place(x=270,y=253)
tk.Button(window,text="     Countdown       ",command=countdown,font=("Helvetica",15)).place(x=270,y=303)
tk.Button(window,text="      Password        ",command=password,font=("Helvetica",15)).place(x=270,y=353)
tk.Button(window,text="      Factorial          ",command=factorial,font=("Helvetica",15)).place(x=270,y=403)
#window show:
window.mainloop()
```
